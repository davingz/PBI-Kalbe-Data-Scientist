## Machine learning regresi (time series)

```python
df_regresi = df_merge.groupby(['Date']).agg({'Qty': 'sum'}).reset_index()
```

```python
df_regresi
```

```python
decomposed = seasonal_decompose(df_regresi.set_index('Date'))

plt.figure(figsize=(8, 20))
plt.subplot(311)
decomposed.trend.plot(ax=plt.gca())
plt.title("Trend")

plt.subplot(312)
decomposed.seasonal.plot(ax=plt.gca())
plt.title('Seasonality')

plt.subplot(313)
decomposed.resid.plot(ax=plt.gca())
plt.title('Residuals')

plt.tight_layout()
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df_regresi['Qty'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
  print('\t%s: %.3f' % (key, value))
```

###### p-value <0.05 artinya data sudah stationer

### check stationarity data

```python
cut_off = round(df_regresi.shape[0] * 0.8)
df_train = df_regresi.iloc[:cut_off]
df_test = df_regresi.iloc[cut_off:].reset_index(drop=True)

df_train.shape, df_test.shape
```

```python
df_train
```

```python
df_test
```

```python
plt.figure(figsize=(20, 5))
sns.lineplot(data=df_train, x=df_train['Date'], y=df_train['Qty'], label='Train Data')
sns.lineplot(data=df_test, x=df_test['Date'], y=df_test['Qty'], label='Test Data')
plt.legend()
plt.show()
```

```python
autocorrelation_plot(df_regresi['Qty']);
```

```python
def rase(y_actual, y_pred):
    # Function to calculate RMSE
    print(f'RMSE value {mean_squared_error(y_actual, y_pred)**0.5}')

def eval (y_actual, y_pred):
    # Function to evaluate machine learning modeling
    rase(y_actual, y_pred)
    print(f'MAE value {mean_absolute_error(y_actual, y_pred)}')
```

```python
#ARIMA

df_train = df_train.set_index('Date')
df_test = df_test.set_index('Date')

y = df_train['Qty']

ARIMAmodel = ARIMA(y, order=(40, 2, 1))
ARIMAmodel = ARIMAmodel.fit()

y_pred = ARIMAmodel.get_forecast(len(df_test))

y_pred_df = y_pred.conf_int()
y_pred_df['predictions'] = ARIMAmodel.predict(start=y_pred_df.index[0], end=y_pred_df.index[-1])
y_pred_df.index = df_test.index
y_pred_out = y_pred_df['predictions']
eval(df_test['Qty'], y_pred_out)

plt.figure(figsize=(20, 9))
plt.plot(df_train['Qty'])
plt.plot(df_test['Qty'], color='red')
plt.plot(y_pred_out, color='black', label='ARIMA Predictions')
plt.legend()
plt.show()
```

### clustering 

```python
df_merge.head()
```

```python
#identifikasi kolom resolusi tinggi
df_merge.corr()
```

```python
df_cluster = df_merge.groupby(['CustomerID']).agg({'TransactionID':'count', 'Qty': 'sum', 'TotalAmount': 'sum'}).reset_index()
```

```python
df_cluster.head()
```

```python
data_cluster = df_cluster.drop(columns=['CustomerID'])
data_cluster_normalize = preprocessing.normalize(data_cluster)
```

```python
data_cluster_normalize
```

```python
df_cluster
```

```python
K = range(2,8)
fits = []
score = []
for k in K:
    model = KMeans(n_clusters = k, random_state = 0, n_init=1).fit(data_cluster_normalize)
    
    fits.append(model)
    
    score.append(silhouette_score(data_cluster_normalize, model.labels_, metric='euclidean'))
```

```python
#chose 4 cluster
sns.lineplot(x = K, y = score);
```

```python
fits[2]
```

```python
df_cluster['cluster label'] = fits[2].labels_
```

```python
df_cluster.groupby(['cluster label']).agg({'CustomerID':'count','TransactionID' : 'mean', 'Qty': 'mean','TotalAmount' : 'mean'})
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Ambil kolom-kolom yang akan digunakan untuk plotting
x = df_cluster['TransactionID'] 
y = df_cluster['Qty'] 
z = df_cluster['TotalAmount']  

# Ambil label kluster dari model K-Means yang telah di-fit
labels = df_cluster['cluster label']  

# Buat plot 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Tentukan warna-warna yang akan digunakan untuk mewarnai kluster
colors = ['r', 'g', 'b', 'c', 'm', 'y']

# Loop melalui setiap label kluster dan plot titik-titiknya dengan warna yang sesuai
for i, label in enumerate(set(labels)):
    x_label = x[labels == label]
    y_label = y[labels == label]
    z_label = z[labels == label]
    ax.scatter(x_label, y_label, z_label, c=colors[i], label=f'Cluster {label}')

# Tambahkan label sumbu
ax.set_xlabel('TransactionID')
ax.set_ylabel('Qty')
ax.set_zlabel('TotalAmount')

# Tambahkan legenda
ax.legend()

# Tampilkan plot
plt.show()
```
